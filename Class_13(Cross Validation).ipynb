# Cross Validation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import LeaveOneOut,KFold,cross_val_score,StratifiedKFold
import pandas as pd
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
my_data = load_iris()
```

```python
features = pd.DataFrame(my_data.data,columns=my_data.feature_names)
features
```

```python
target = pd.DataFrame(my_data.target,columns=['Target'])
target
```

## Leaveone out

```python
leave = LeaveOneOut()
leave.split(features,target)
```

```python
for train_split,test_split in leave.split(features,target):
    print(train_split,test_split)
```

## KFold

```python
kfold = KFold(n_splits=5,shuffle=True)
```

```python
for train_split,test_split in kfold.split(features,target):
    print(train_split,test_split)
```

## StratifiedKFold

```python
stra_kfold = KFold(n_splits=5,shuffle=True)
```

```python
for train_split,test_split in stra_kfold.split(features,target):
    print(train_split,test_split)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cross_val_score(LogisticRegression(),features,target,scoring='accuracy',cv=5).mean()
#cv=Meantion any method from 3 of them or else meantion a number
```
