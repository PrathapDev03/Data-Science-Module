# Model Deployment

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import streamlit as st
```

```python
df = pd.read_csv('claimants.csv')
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
log_model = LogisticRegression()
```

```python
target = df[['ATTORNEY']]
target.head()
```

```python
features = df.drop(columns=['CASENUM','ATTORNEY'])
features.head()
```

```python
log_model.fit(features,target)
```

```python
y_pred = log_model.predict(features)
```

```python
accuracy_score(target,y_pred)
```
