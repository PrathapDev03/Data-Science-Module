# Data Cleaning

```python
# Data Cleaning
def data_clean(words):
  data=' '.join(re.findall('\w+',words))
  doc=nlp(data)
  clean_text=[token.lemma_ for token in doc if not token.is_stop and not token.is_punct and not token.is_bracket and not token.is_digit and not token.is_currency]
  return clean_text
```

```python
for i,j in count.vocabulary_.items():
  print(i,':-',sum_x[0,j])
```

```python
Noun_verbs = [(i,sum_x[0,j]) for i,j in count.vocabulary_.items()]
Noun_verb = sorted(Noun_verbs,key=lambda x:x[1],reverse=True)
words_df = pd.DataFrame(Noun_verb,columns=['words','count'])
words_df
```
