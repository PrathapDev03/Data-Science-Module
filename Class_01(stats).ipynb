# Z-Test 

```python
1.The average height of all residents in a city is 168cm with a population std = 3.9.A doctor belives the 
mean is different.He measured the height of 36 individuals and found the average to be 169.5cm

a)State Null and Alternative hypothesis.
b)At a 95%,CI is there enough evidence to reject the Null Hypothesis
```

```python
# H0 = pop_mean = 168cm 
# Ha = pop_mean!= 168cm
# Given Data
pop_mean = 168
pop_std_dev = 3.9
n = 36
x_bar = 169.5
# CI = 95%(0.95),alpha=5%(0.05)
```

```python
# Z_test = (x_bar - pop_mean)/(pop_std_dev/sqrt(n))
```

```python
Z_test = (x_bar - pop_mean)/(pop_std_dev/np.sqrt(n))
Z_test
```

```python
# z_critical = 1.96
# since s_critical < z_score reject H0
# pop_mean!=168
```

```python
z_critical = st.norm.ppf(0.975)       # ppf = point percentile field
z_critical
```

```python
2.Scores on  an exam are normally distributed with a population std_dev = 5.6.A random sample of 40 scores on 
the exam has a mean of 32.Estimate the population mean with 
a)80% confidence
b)90% confidence
c)98% confidence
```

```python
# Given data
pop_std_dev = 5.6
n = 40
x_bar = 32
```

```python
# pop_mean_with_80%_confidence
z_critical_80 = stats.norm.ppf(0.90)
print('z_critical is :',z_critical)
lower_limit_80 = x_bar-(z_critical_80*(pop_std_dev/np.sqrt(n)))
print('Lower limit is :',lower_limit_80)
upper_limit_80 = x_bar+(z_critical_80*(pop_std_dev/np.sqrt(n)))
print('Upper limit is :',upper_limit_80)
```

```python
# pop_mean_with_90%_confidence
z_critical_90 = stats.norm.ppf(0.95)
print('z_critical is :',z_critical)
lower_limit_90 = x_bar-(z_critical_90*(pop_std_dev/np.sqrt(n)))
print('Lower limit is :',lower_limit_90)
upper_limit_90 = x_bar+(z_critical_90*(pop_std_dev/np.sqrt(n)))
print('Upper limit is :',upper_limit_90)
```

```python
# pop_mean_with_98%_confidence
z_critical_98 = stats.norm.ppf(0.99)
print('z_critical is :',z_critical)
lower_limit_98 = x_bar-(z_critical_98*(pop_std_dev/np.sqrt(n)))
print('Lower limit is :',lower_limit_98)
upper_limit_98 = x_bar+(z_critical_98*(pop_std_dev/np.sqrt(n)))
print('Upper limit is :',upper_limit_98)
```

### 2 tail independent tests:

```python
A statistics teacher wants to compare his two classes to see if they performed any differently on the tests 
he gave that semester. Class A had 25 students with an average score of 70, standard deviation 15. 
Class B had 20 students with an average score of 74, standard deviation 25. Using alpha 0.05 did these 
two classes perform differently on the tests?.
```

```python
# H0 = Two classes performance are same
# Ha = Two classes performance are not same
```

```python
# Given data
# Class A
n1=25
x1_bar=70
s1=15
#class B
n2=20
x2_bar=74
s2=25
#alpha = 0.05 , CI = 0.95(95%)
df1=n1-1
df2=n2-1
Total_df = df1+df2
Total_df
```

```python
sp_sqr = (((s1**2)*df1) + ((s2**2)*df2)) / Total_df
sp_sqr
```

```python
t_stats = (x1_bar - x2_bar) / np.sqrt((sp_sqr/n1)+(sp_sqr/n2))
t_stats
```

```python
t_critical=stats.t.ppf(0.025,43)
t_critical
```

```python
since tstats>tcritical fail to reject H0
Two class performance are same
```

```python
sample1 = stats.norm.rvs(size=25,loc=70,scale=15)
sample2 = stats.norm.rvs(size=20,loc=74,scale=25)
```

# Paired Test

```python
A study has conducted to determine the effectiveness of a weight base program. The table below shows the before 
and after weights of 10 subject in the program. 
A) Is the program effective for reducing weight? 
B) (use a 5% significant level) construct a 95% confidence interval and determine the margin of error.
```

```python
s.no   | Before | After |
1         185     169
2         192     187
3         206     193
4         177     176
5         225     194
6         168     171
7         256     228
8         239     217
9         199     204
10        218     195
```

```python
# H0 = Program is not effective
# Ha = Program is effective
```

```python
# Given data
n=10
# Alpha = 0.05 , CI = 0.95
```

```python
Before = [185,192,206,177,225,168,256,239,199,218]
After = [169,187,193,176,194,171,228,217,204,195]
```

```python
# Tstats = d_bar / (sd_bar/np.sqrt(n))
```

```python
diff = []
for i in range(len(After)):
    diff.append(After[i]-Before[i])
diff
```

```python
d_bar = np.mean(diff)
d_bar
```

```python
sd_bar = np.std(diff,ddof=1)
sd_bar
```

```python
tstats = d_bar / (sd_bar/np.sqrt(n))
tstats
```

```python
# Package
stats.ttest_rel(After,Before)
```

```python
t_critical=stats.t.ppf(0.025,9)
t_critical
```

```python
since tstats < tcritical is rejected
```

# ANOVA Test

```python
a = [2,3,7,2,6]
b = [10,8,7,5,10]
c = [10,13,14,13,15]
```

```python
#Fstats = (variance_b/w_group/df1) / (variance_within_group/df2)
```

```python
# variance_b/w_group
mean_a = np.mean(a)
mean_b = np.mean(b)
mean_c = np.mean(c)
print(mean_a,mean_b,mean_c)
```

```python
overall_mean = (mean_a+mean_b+mean_c)/3
overall_mean
```

```python
n=5
```

```python
variance_bw_group = n*((overall_mean-mean_a)**2)+n*((overall_mean-mean_b)**2)+n*((overall_mean-mean_c)**2)
variance_bw_group
```

```python
variance_within_group = sum((a-mean_a)**2)+sum((b-mean_b)**2)+sum((c-mean_c)**2)
variance_within_group
```

```python
df1 = 2
df2 = 12
```

```python
Fstats = (variance_bw_group/df1) / (variance_within_group/df2)
Fstats
```

```python
stats.f_oneway(a,b,c)
```

```python
F_critical = stats.f.ppf(0.95,2,12)
F_critical
```
